#### D問題

条件
- ¥640の海鮮丼
- 節分の日(02/02)には恵方巻きを¥540で提供
- 「9のつく日」には「9の個数×30円」だけ海鮮丼も恵方巻も値引きになる。
    （たとえば1996年2月29日の海鮮丼は550円）
- 対象期間は1979年〜2021年

対象期間中、海鮮丼1つを毎日購入し続けた（節分の日には恵方巻も追加）場合の合計金額は？

```python
import datetime

# -----------------------------------------------------------------------------
# 条件
KAISENDON_PRICE = 640
EHOUMAKI_PRICE = 540
# 特別メニュー提供の日（今回は節分の日）
SETSUBUNL_DAY = '0202'
# 期間設定
START_YEAR = 1979
END_YEAR = 2021


def create_date_list_by_year(start_year, end_year):
    # 開始日と終了日を datetime.date オブジェクトに設定
    start_date = datetime.date(start_year, 1, 1)
    end_date = datetime.date(end_year, 12, 31)
    
    # 日付のリストを生成
    date_list = []
    current = start_date
    while current <= end_date:
        date_list.append(current.strftime("%Y%m%d"))
        current += datetime.timedelta(days=1)
    
    return date_list


date_lst = create_date_list_by_year(START_YEAR, END_YEAR)


# -----------------------------------------------------------------------------
# 演算
total_price = 0

for date in date_lst:
    day_total = 0

    # 割引額
    if date.count('9') > 0:
        discount_price = date.count('9') * 30
    else:
        discount_price = 0

    if date.endswith(SETSUBUNL_DAY):
        day_total = KAISENDON_PRICE + EHOUMAKI_PRICE - discount_price * 2
    else:
        day_total = KAISENDON_PRICE - discount_price

    total_price += day_total


# output
print(f'{START_YEAR}年〜{END_YEAR}年における丼丸での購入金額の合計は¥{total_price:,}')
```
